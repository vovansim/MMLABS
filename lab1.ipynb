# 1)**x'=-kx** 
Модель мальтуса

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(y, t):
    k = 0.42
    return -k * y

t = np.linspace(0, 10, 100)
print(t)
```

```python
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()

%matplotlib inline

plt.plot(t, solve, label = r"$x'= -kx$")
plt.grid(True)
plt.legend()
plt.show()
```

# 2)**x'=k(x)x**
Мягкая модель роста с коэффициентом размножения, зависящим от населения

```python
def k(x):
    a = 0.42
    b = 0.31
    return a - b * x
def dydt2(x, t):
    return k(x) * x

solve = odeint(dydt2, y0=1, t=t)
solve = solve.flatten()

%matplotlib inline

plt.plot(t, solve, label = r"$x'= k(x)x$")
plt.grid(True)
plt.legend()
plt.show()
```

# 3)**x'=x-x^2-c**
Модель изменения популяции рыб при рыболовстве с интенсивностью (с=-0.42)

```python
def dydt3(x, t):
    c = - 0.42
    return x - x ** 2 - c
solve = odeint(dydt3, y0=1, t=t)
solve = solve.flatten()
%matplotlib inline

plt.plot(t, solve, label = r"$x'= x-x^2-c$")
plt.grid(True)
plt.legend()
plt.show()
```

# 4) **x'=x-g(x)x-c**
Мягкая модель изменения популяции рыб

```python
def g(x):
    a = 0.6
    b = 0.0009
    return a - b * x
def dydt4(x, t):
    c = 0.7
    return x - g(x) * x - c
solve = odeint(dydt4, y0=1, t=t)
solve = solve.flatten()
%matplotlib inline

plt.plot(t, solve, label = r"$x'= x-g(x)x-c$")
plt.grid(True)
plt.legend()
plt.show()
```

# **x'=x-x^2-kx**
Модель изменения популяции рыб при дифференциальной квоте вылова

```python
def dxdt5(x, t):
    k = 0.7
    return x - x ** 2 - k * x
solve = odeint(dxdt5, y0=1, t=t)
solve = solve.flatten()
%matplotlib inline

plt.plot(t, solve, label = r"$x'= x-x^2-kx$", color = "blue")
plt.grid(True)
plt.legend()
plt.show()
```
