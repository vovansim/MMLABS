```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

h=0.15

def dydt(y, t):
    k = 0.42
    return -k * y

def Euler(f, y0, t, h):
    y=[]
    y.append(y0)
    for i in range(1,len(t)):
        yn=y[i-1]+h*f(y[i-1],t[i-1])
        y.append(yn)
    return y



t = np.linspace(0, 10, 100)
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
euler=Euler(dydt,1,t,h)

%matplotlib inline

plt.plot(t, euler, label = r"euler$")
plt.plot(t, solve, label = r"$x'= -kx$")
plt.grid(True)
plt.legend()
plt.show()
```
