*y' = -ky*

```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy.integrate import odeint
```

```python
def dydt (y, t):
    k = 0.42
    return -k * y
```

```python
t = np.linspace(0, 10, 100)
```

```python
solve = odeint (dydt, y0=1, t=t)
solve = solve.flatten()
```

```python
%matplotlib inline

plt.plot(t, solve, label = r"$y' = -ky$")
plt.grid(True)
plt.legend()
plt.show()
print (t)
```
